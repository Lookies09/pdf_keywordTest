```python
import os
import sys

sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))

from src.core.pdf_engine import extract_text, analyze_keywords

# 테스트용 로직 실행
raw_path = "../data/raw"
files = sorted([f for f in os.listdir(raw_path) if f.endswith('.pdf')])

trend_result = {}

for file in files:
    year = file.split('.')[0]
    file_full_path = os.path.join(raw_path, file)
    
    text = extract_text(file_full_path)
    keywords = analyze_keywords(text, top_n=50)
    
    trend_result[year] = {word: count for word, count in keywords}

# 결과 확인
import json
print(json.dumps(trend_result, indent=2, ensure_ascii=False))
```
